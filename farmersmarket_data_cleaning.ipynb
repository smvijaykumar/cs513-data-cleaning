```python
import numpy as np
import pandas as pd
from datetime import datetime
from IPython.display import display
import sqlite3 as sql
```

```python
# Load data from farmermarkets listings csv file
farmers_df = pd.read_csv("farmersmarkets_imputed.csv")
```

```python
def trim_head_tail_space(df_column):
    clean_df_column = df_column.str.strip()
    print("row changes in column " + str(df_column.name) +": " ,sum(df_column!=clean_df_column))
    return clean_df_column

### Run function
farmers_df['MarketName'] = trim_head_tail_space(farmers_df['MarketName'])
farmers_df['Location'] = trim_head_tail_space(farmers_df['Location'])
farmers_df['Website'] = trim_head_tail_space(farmers_df['Website'])
farmers_df['Facebook'] = trim_head_tail_space(farmers_df['Facebook'])
farmers_df['Twitter'] = trim_head_tail_space(farmers_df['Twitter'])
farmers_df['Youtube'] = trim_head_tail_space(farmers_df['Youtube'])
farmers_df['OtherMedia'] = trim_head_tail_space(farmers_df['OtherMedia'])
farmers_df['street'] = trim_head_tail_space(farmers_df['street'])
farmers_df['Season1Date_Start'] = trim_head_tail_space(farmers_df['Season1Date_Start'])
farmers_df['Season1Date_End'] = trim_head_tail_space(farmers_df['Season1Date_End'])
```

```python
### Define function
def remove_consecutive_spaces(df_column):
    clean_df_column = df_column.replace('\s\s+', ' ', regex=True)
    print("row changes in column " + str(df_column.name) +": " ,sum(df_column!=clean_df_column))
    return clean_df_column

### Run function
farmers_df['MarketName'] = remove_consecutive_spaces(farmers_df['MarketName'])
farmers_df['Location'] = remove_consecutive_spaces(farmers_df['Location'])
farmers_df['Website'] = remove_consecutive_spaces(farmers_df['Website'])
farmers_df['Facebook'] = remove_consecutive_spaces(farmers_df['Facebook'])
farmers_df['Twitter'] = remove_consecutive_spaces(farmers_df['Twitter'])
farmers_df['Youtube'] = remove_consecutive_spaces(farmers_df['Youtube'])
farmers_df['OtherMedia'] = remove_consecutive_spaces(farmers_df['OtherMedia'])
farmers_df['street'] = remove_consecutive_spaces(farmers_df['street'])
farmers_df['Season1Date_Start'] = remove_consecutive_spaces(farmers_df['Season1Date_Start'])
farmers_df['Season1Date_End'] = remove_consecutive_spaces(farmers_df['Season1Date_End'])
```

```python
### Define function
def remove_special_characters(df_column,bad_characters_list):
    clean_df_column = df_column
    for bad_char in bad_characters_list:
        clean_df_column = clean_df_column.str.replace(bad_char,' ')
        print("row changes in column " + str(df_column.name) + " after removing character " + str(bad_char) + ": " ,sum(df_column!=clean_df_column))
    clean_df_column = clean_df_column.str.title()
    return clean_df_column

### Run function
bad_chars_lst = ["*","!","?", "(", ")", "-"]
farmers_df['MarketName_clean'] = remove_special_characters(farmers_df['MarketName'],bad_chars_lst)
display(farmers_df[["MarketName","MarketName_clean"]].head(20))
```

```python
### Identify columns with too many missing values
for col in farmers_df.columns:
    percentage_missing = np.mean(farmers_df[col].isnull())
    print('{} - {}%'.format(col, round(percentage_missing*100)))
```

```python
for col in farmers_df.columns:
    missing = farmers_df[col].isnull()
    num_missing = np.sum(missing)
    if num_missing > 0:  
        farmers_df['{}_ismissing'.format(col)] = missing

### Based on the indicator, count number of Nan values in each row
ismissing_cols = [col for col in farmers_df.columns if 'ismissing' in col]
farmers_df['num_missing'] = farmers_df[ismissing_cols].sum(axis=1)

### Remove rows with number of Nan values more than n (n=3)
ind_missing = farmers_df[farmers_df['num_missing'] >= 3].index
df_less_missing_rows = farmers_df.drop(ind_missing, axis=0)

### Print out some examples of rows with many missing values. 
### Show overall changes in your csv. 

print ("Number of rows BEFORE cleaning missing values: ", farmers_df.shape[0])
print ("Number of rows AFTER cleaning missing values: ", df_less_missing_rows.shape[0])
```

```python
def replace_month_with_startdate(df_column):
    clean_df_column = df_column.replace("January", "01/01/10")
    clean_df_column = clean_df_column.replace("February", "02/01/10")
    clean_df_column = clean_df_column.replace("March", "03/01/10")
    clean_df_column = clean_df_column.replace("April", "04/01/10")
    clean_df_column = clean_df_column.replace("May", "05/01/10")
    clean_df_column = clean_df_column.replace("June", "06/01/10")
    clean_df_column = clean_df_column.replace("July", "07/01/10")
    clean_df_column = clean_df_column.replace("August", "08/01/10")
    clean_df_column = clean_df_column.replace("September", "09/01/10")
    clean_df_column = clean_df_column.replace("October", "10/01/10")
    clean_df_column = clean_df_column.replace("Oc", "10/01/10")
    clean_df_column = clean_df_column.replace("November", "11/01/10")
    clean_df_column = clean_df_column.replace("December", "12/01/10")
    return clean_df_column

def replace_month_with_enddate(df_column):
    clean_df_column = df_column.replace("January", '01/31/2010')
    clean_df_column = clean_df_column.replace("February", '02/28/2010')
    clean_df_column = clean_df_column.replace("March", '03/31/2010')
    clean_df_column = clean_df_column.replace("April", '04/30/2010')
    clean_df_column = clean_df_column.replace("May", '05/31/2010')
    clean_df_column = clean_df_column.replace("June", '06/30/2010')
    clean_df_column = clean_df_column.replace("July", '07/31/2010')
    clean_df_column = clean_df_column.replace("August", '08/31/2010')
    clean_df_column = clean_df_column.replace("September", '09/30/2010')
    clean_df_column = clean_df_column.replace("October", '10/31/2010')
    clean_df_column = clean_df_column.replace("Oc", '10/31/2010')
    clean_df_column = clean_df_column.replace("November", '11/30/2010')
    clean_df_column = clean_df_column.replace("December", '12/31/2010')
    return clean_df_column

farmers_df["Season1Date_Start"] = replace_month_with_startdate(farmers_df["Season1Date_Start"])
farmers_df["Season1Date_End"]= replace_month_with_enddate(farmers_df["Season1Date_End"])
```

```python
# converting the string to datetime
# format in multiple columns
farmers_df['Season1Date_Start'] = pd.to_datetime(
                          farmers_df['Season1Date_Start'],
                          dayfirst=False, errors='coerce')
farmers_df['Season1Date_End'] = pd.to_datetime(
                          farmers_df['Season1Date_End'],
                          dayfirst=False, errors='coerce')
```
